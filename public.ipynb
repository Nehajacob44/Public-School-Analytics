```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=df, x='CATEGORY')
plt.show()
```

```python
df.head()
```

```python
city_counts = df['CITY'].value_counts().head(5) 
plt.figure(figsize=(10, 6))
sns.barplot(x=city_counts.index, y=city_counts.values)
plt.xlabel('City')
plt.ylabel('Number of Schools')
plt.title('Number of Schools in Each City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
s =df[df['CITY'] == 'Silver Spring']
category_counts = s['CATEGORY'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(category_counts.index, category_counts.values)
plt.title('School Category Distribution in Silver Spring', fontsize=14)
plt.xlabel('School Category', fontsize=12)
plt.ylabel('Number of Schools', fontsize=12)
plt.tight_layout()
```
